```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

def add_box(ax, text, x, y, width, height):
    box = FancyBboxPatch((x, y), width, height,
                         boxstyle="round,pad=0.3",
                         edgecolor='black',
                         facecolor='lightgrey',
                         linewidth=1.5)
    ax.add_patch(box)
    ax.text(x + width / 2, y + height / 2, text, ha='center', va='center', fontsize=10)

def add_arrow(ax, x1, y1, x2, y2, label=None):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(facecolor='black', shrink=0.05, width=0.5, headwidth=5))
    if label:
        ax.text((x1 + x2) / 2, (y1 + y2) / 2, label, ha='center', va='center', fontsize=9, color='black')

def draw_flowchart():
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.set_xlim(-5, 5)
    ax.set_ylim(-15, 5)
    ax.axis('off')

    # Adding boxes for browser, proxy server, and server
    add_box(ax, "浏览器", -4.5, 0, 1.5, 1)
    add_box(ax, "代理服务器", -1, 0, 2, 1)
    add_box(ax, "原服务器", 2.5, 0, 1.5, 1)

    # Adding flow steps
    add_box(ax, "Proxysocket()", -1, -2, 2, 1)
    add_box(ax, "Bind()", -1, -3.5, 2, 1)
    add_box(ax, "Listen()", -1, -5, 2, 1)
    add_box(ax, "Accept()", -1, -6.5, 2, 1)
    add_box(ax, "Recvfrom()\n(从客户端)", -1, -8, 2, 1.5)
    add_box(ax, "Serversocket()", -1, -10, 2, 1)
    add_box(ax, "Connect()", -1, -11.5, 2, 1)
    add_box(ax, "Sendto()\n(到服务器)", -1, -13, 2, 1.5)
    add_box(ax, "Recvfrom()\n(从服务器)", -1, -15, 2, 1.5)
    add_box(ax, "Sendto()\n(到客户端)", -1, -17, 2, 1.5)
    add_box(ax, "close()//关闭所有连接", -1, -19, 2, 1.5)

    # Adding arrows between the boxes
    add_arrow(ax, -1, -1, -1, -2)
    add_arrow(ax, -1, -2.5, -1, -3.5)
    add_arrow(ax, -1, -4, -1, -5)
    add_arrow(ax, -1, -5.5, -1, -6.5)
    add_arrow(ax, -1, -7.5, -1, -8)
    add_arrow(ax, -1, -9.5, -1, -10)
    add_arrow(ax, -1, -10.5, -1, -11.5)
    add_arrow(ax, -1, -12.5, -1, -13)
    add_arrow(ax, -1, -14.5, -1, -15)
    add_arrow(ax, -1, -16.5, -1, -17)
    add_arrow(ax, -1, -18.5, -1, -19)

    plt.show()

draw_flowchart()
```
